## Load 2 maps and 2 weight maps

```python
nside = 32
ellmax = 50
wtmap1_freq, wtmap1_scale, wtmap2_freq, wtmap2_scale  =0, 0, 0, 0 #works for i,1,j,1
p, q = 1,1 #0 for wt maps with preserved CMB, 1 for preserved tSZ
y, z = 0,0 #0 for CMB, 1 for tSZ
band_limit = True

wtmap1_preserved = 'CMB' if p==0 else 'tSZ'
wtmap2_preserved = 'CMB' if q==0 else 'tSZ'

ells = np.arange(ellmax+1)
input_dir = '/Users/kristen/Documents/NILC_plots/inputs'
output_dir = '/Users/kristen/Documents/NILC_plots/outputs'


if y==0:
    map1 = hp.read_map(f'{input_dir}/cmb_lensed_nside1024_K.fits')
#     map1_cl = hp.anafast(map1)
#     map1 = hp.synfast(map1_cl, nside)
else:
    map1 = 1000*hp.read_map(f'{input_dir}/halosky_maps/tsz_00000.fits')
if z==0:
    map2 = hp.read_map(f'{input_dir}/cmb_lensed_nside1024_K.fits')
#     map2_cl = hp.anafast(map2)
#     map2 = hp.synfast(map2_cl, nside)
else:
    map2 = 1000*hp.read_map(f'{input_dir}/halosky_maps/tsz_00000.fits')
wtmap1 = hp.read_map(f'{output_dir}/pyilc_outputs/sim0weightmap_freq{wtmap1_freq}_scale{wtmap1_scale}_component_{wtmap1_preserved}.fits')
if p==0:
    wtmap1 *= 10**(-6)
wtmap2 = hp.read_map(f'{output_dir}/pyilc_outputs/sim0weightmap_freq{wtmap2_freq}_scale{wtmap2_scale}_component_{wtmap2_preserved}.fits')
if q==0:
    wtmap2 *= 10**(-6)
    
map1 = hp.ud_grade(map1, nside)
map2 = hp.ud_grade(map2, nside)
wtmap1 = hp.ud_grade(wtmap1, nside)
wtmap2 = hp.ud_grade(wtmap2, nside)
```

## Band-limiting

```python
if band_limit:
    
    lmax_data = 3*nside-1
    l_arr,m_arr = hp.Alm.getlm(lmax_data)
    alm1 = hp.map2alm(map1)
    alm2 = hp.map2alm(map2)
    wlm1 = hp.map2alm(wtmap1)
    wlm2 = hp.map2alm(wtmap2)
    alm1 = alm1*(l_arr<=ellmax)
    alm2 = alm2*(l_arr<=ellmax)
    wlm1 = wlm1*(l_arr<=ellmax)
    wlm2 = wlm2*(l_arr<=ellmax)

    map1 = hp.alm2map(alm1, nside=nside)
    map2 = hp.alm2map(alm2, nside=nside)
    wtmap1 = hp.alm2map(wlm1, nside=nside)
    wtmap2 = hp.alm2map(wlm2, nside=nside)
```

## Directly Compute Power Spectrum of Two Weighted Maps

```python
masked_map1 = map1*wtmap1
masked_map2 = map2*wtmap2
directly_computed = hp.anafast(masked_map1, masked_map2, lmax=ellmax)
```

## reMASTERed Computation of Cross-Spectrum of Two Weighted Maps

```python
def calculate_all_cl(wigner, h, g, Clzz, Clw1w2, Clzw, w, a,
    bispectrum_zzw, bispectrum_wzw, Rho, delta_ij=False):
    '''
    Does vectorized power spectrum calculation for one ell

    ARGUMENTS
    ---------
    inp: Info object, contains input specifications
    h: 2D list of needlet filter function values, indexed as h[n][l] where n is needlet filter scale
    g: list of g_i spectral response at different frequencies, indexed as g[i]
    Clzz: 1D numpy array of size ell_sum_max containing component power spectrum
    Clw1w2: index as Clw1w2[n,m,i,j,l] to get cross spectra of weight map pi(n) and qj(m)
    Clzw: index as Clzw[n,i,l]
    w: indexed as w[n,i], gives means of weight maps
    a: mean of map
    bispectrum_zzw: indexed as bispectra[m,j,b1,b2,b3]
    bispectrum_wzw: indexed as bispectra[n,i,m,j,b1,b2,b3]
    Rho: indexed as rho[n,i,m,j,b2,b4,b3,b5,b1]
    delta_ij: True if delta_{ij} is attached to term. False by default

    RETURNS
    -------
    Cl: (N_preserved_comps, N_preserved_comps, ell_max) numpy array of floats 
        containing value of power spectrum from one component's contribution
        at each ell through ellmax, index as Cl[p,q,l]

    INDICES MAPPING IN EINSUM
    -------------------------
    l1, l2, l3, l4, l5 -> l, a, b, c, d
    b1, b2, b3, b4, b5 -> e, f, g, h, k

    '''
    dl_bispectrum = 1
    dl_trispectrum = 10
    ell_sum_max = 50

    wigner = wigner[:51, :51, :51]
    l1 = np.arange(51)
    l2 = np.arange(51)
    l3 = np.arange(51)

    #Define theta functions where ell_bins[b,l]=1 if l in bin b, 0 otherwise
    ells_sum = np.arange(51)
    Nl_sum_bispec = ell_sum_max//dl_bispectrum
    ell_sum_bins_bispec = [(ells_sum>=dl_bispectrum*bin1)&(ells_sum<dl_bispectrum*(bin1+1)) for bin1 in range(Nl_sum_bispec)]
    ell_sum_bins_bispec[-1][-1] = 1 #put highest ell value in last bin
    Nl_sum_trispec = ell_sum_max//dl_trispectrum
    ell_sum_bins_trispec = [(ells_sum>=dl_trispectrum*bin1)&(ells_sum<dl_trispectrum*(bin1+1)) for bin1 in range(Nl_sum_trispec)]
    ell_sum_bins_trispec[-1][-1] = 1 #put highest ell value in last bin

    ells = np.arange(ellmax+1)
    Nl_bispec = ellmax//dl_bispectrum
    ell_bins_bispec = [(ells>=dl_bispectrum*bin1)&(ells<dl_bispectrum*(bin1+1)) for bin1 in range(Nl_bispec)]
    ell_bins_bispec[-1][-1] = 1 #put highest ell value in last bin
    Nl_trispec = ellmax//dl_trispectrum
    ell_bins_trispec = [(ells>=dl_trispectrum*bin1)&(ells<dl_trispectrum*(bin1+1)) for bin1 in range(Nl_trispec)]
    ell_bins_trispec[-1][-1] = 1 #put highest ell value in last bin

    if not delta_ij:
        aa_ww_term = float(1/(4*np.pi))*np.einsum('a,b,lab,lab,a,b->l',2*l2+1,2*l3+1,wigner[:ellmax+1,:,:],wigner[:ellmax+1,:,:],Clzz,Clw1w2,optimize=True)
        aw_aw_term = float(1/(4*np.pi))*np.einsum('a,b,lab,lab,a,b->l',2*l2+1,2*l3+1,wigner[:ellmax+1,:,:],wigner[:ellmax+1,:,:],Clzw1,Clzw2,optimize=True)
        w_aaw_term = float(1/(4*np.pi))*1/(2*np.sqrt(np.pi))*w1*np.einsum('a,b,lab,lab,efg,el,fa,gb->l',2*l2+1,2*l3+1,wigner[:ellmax+1,:,:],wigner[:ellmax+1,:,:],bispectrum_zzw2,ell_bins_bispec,ell_sum_bins_bispec,ell_sum_bins_bispec,optimize=True) \
                    + float(1/(4*np.pi))*1/(2*np.sqrt(np.pi))*w2*np.einsum('a,b,lab,lab,efg,el,fa,gb->l',2*l2+1,2*l3+1,wigner[:ellmax+1,:,:],wigner[:ellmax+1,:,:],bispectrum_zzw1,ell_bins_bispec,ell_sum_bins_bispec,ell_sum_bins_bispec,optimize=True)
        a_waw_term = float(1/(4*np.pi))*1/(2*np.sqrt(np.pi))*a*np.einsum('a,b,lab,lab,efg,el,fa,gb->l',2*l2+1,2*l3+1,wigner[:ellmax+1,:,:],wigner[:ellmax+1,:,:],bispectrum_wzw,ell_bins_bispec,ell_sum_bins_bispec,ell_sum_bins_bispec,optimize=True)
        aaww_term = np.einsum('l,fhgke,el,fa,gb,hc,kd->l', 1/(2*l1+1), Rho,ell_bins_trispec,ell_sum_bins_trispec,ell_sum_bins_trispec,ell_sum_bins_trispec,ell_sum_bins_trispec, optimize=True)
        
    Cl = aa_ww_term + aw_aw_term + w_aaw_term + a_waw_term + aaww_term
    return Cl, np.array([aa_ww_term, aw_aw_term, w_aaw_term, a_waw_term, aaww_term])
```

```python
Clzz = pickle.load(open(f'{output_dir}/n_point_funcs/sim0_Clzz.p', 'rb'))
Clw1w2 = pickle.load(open(f'{output_dir}/n_point_funcs/sim0_Clw1w2.p', 'rb'))
Clzw = pickle.load(open(f'{output_dir}/n_point_funcs/sim0_Clzw.p', 'rb'))
a = pickle.load(open(f'{output_dir}/n_point_funcs/sim0_a.p', 'rb'))
w = pickle.load(open(f'{output_dir}/n_point_funcs/sim0_w.p', 'rb'))
bispectrum_zzw = pickle.load(open(f'{output_dir}/n_point_funcs/sim0_bispectrum_zzw.p', 'rb'))
bispectrum_wzw = pickle.load(open(f'{output_dir}/n_point_funcs/sim0_bispectrum_wzw.p', 'rb'))
Rho = pickle.load(open(f'{output_dir}/n_point_funcs/sim0_Rho.p', 'rb'))

#index n_point funcs for z=1, p=q=1, n=wtmap1_scale, m=wtmap2_scale, i=wtmap1_freq, j=wtmap2_freq
Clzz = Clzz[z] #index as l
Clw1w2 = Clw1w2[p,q,wtmap1_scale,wtmap2_scale,wtmap1_freq,wtmap2_freq] #index as l
Clzw1 = Clzw[z,p,wtmap1_scale,wtmap1_freq] #index as l
Clzw2 = Clzw[z,q,wtmap2_scale,wtmap2_freq] #index as l
a = a[z] #no indices
w1 = w[p,wtmap1_scale,wtmap1_freq] #no indices
w2 = w[q,wtmap2_scale,wtmap2_freq] #no indices
bispectrum_zzw1 = bispectrum_zzw[z,p,wtmap1_scale,wtmap1_freq] #index as b1 b2 b3
bispectrum_zzw2 = bispectrum_zzw[z,q,wtmap2_scale,wtmap2_freq] #index as b1 b2 b3
bispectrum_wzw = bispectrum_wzw[p,wtmap1_scale,wtmap1_freq,z,q,wtmap2_scale,wtmap2_freq] #index as b1 b2 b3
Rho = Rho[z,p,wtmap1_scale,wtmap1_freq,q,wtmap2_scale,wtmap2_freq] #index as b2 b4 b3 b5 b1
```

```python
Cl, terms = calculate_all_cl(wigner, h, g, Clzz, Clw1w2, Clzw, w, a,
    bispectrum_zzw, bispectrum_wzw, Rho, delta_ij=False)
aa_ww, aw_aw, w_aaw, a_waw, aaww = terms
```

```python
def GaussianNeedlets(ELLMAX, FWHM_arcmin=np.array([30.])):
    '''
    Function from pyilc (https://github.com/jcolinhill/pyilc)

    ARGUMENTS
    ---------
    ELLMAX: int, maximum ell for needlet filters
    FWHM_arcmin: array of FWHM used for constrution of Gaussians (needlet filters
        are differences of two Gaussians). FWHM need to be in strictly decreasing 
        order, otherwise you'll get nonsense


    RETURNS
    --------
    ell: array of ell values
    filters: (N_scales, ellmax+1) numpy array containing filters at each scale

    '''
    if ( any( i <= j for i, j in zip(FWHM_arcmin, FWHM_arcmin[1:]))):
        raise AssertionError
    ell = np.arange(ELLMAX+1)
    N_scales = len(FWHM_arcmin) + 1
    filters = np.zeros((N_scales, ELLMAX+1))
    FWHM = FWHM_arcmin * np.pi/(180.*60.)
    # define gaussians
    Gaussians = np.zeros((N_scales-1, ELLMAX+1))
    for i in range(N_scales-1):
        Gaussians[i] = hp.gauss_beam(FWHM[i], lmax=ELLMAX)
    # define needlet filters in harmonic space
    filters[0] = Gaussians[0]
    for i in range(1,N_scales-1):
        filters[i] = np.sqrt(Gaussians[i]**2. - Gaussians[i-1]**2.)
    filters[N_scales-1] = np.sqrt(1. - Gaussians[N_scales-2]**2.)
    # simple check to ensure that sum of squared transmission is unity as needed for NILC algorithm
    assert (np.absolute( np.sum( filters**2., axis=0 ) - np.ones(ELLMAX+1,dtype=float)) < 1.e-3).all(), "wavelet filter transmission check failed"
#     taper_width = 5.
#     taper_func = (1.0 - 0.5*(np.tanh(0.025*(ell - (ELLMAX - taper_width))) + 1.0)) #smooth taper to zero from ELLMAX-taper_width to ELLMAX
    # taper_func *= 0.5*(np.tanh(0.5*(ell-10)))+0.5 #smooth taper to zero for low ell
    # for i, filt in enumerate(filters):
    #     filters[i] = filters[i]*taper_func
    return ell, filters

ell, filters = GaussianNeedlets(ellmax)
```
